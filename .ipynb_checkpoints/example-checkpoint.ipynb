```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import arma
import nn
import armann
import armann_zhang

import addmet
```

```python
#Data
AMZN = pd.read_csv(r"Data/AMZN.csv")
RUB = pd.read_csv(r"Data/RUB=X.csv")
SNP = pd.read_csv(r"Data/^GSPC.csv")
amzn_close = np.array(AMZN.Close[259:])
rub_close = np.array(RUB.Close[303:])
index = np.argwhere(rub_close=="null")
rub_close = np.delete(rub_close, index)
rub_close = np.float64(rub_close)
amzn_close_diff = np.diff(amzn_close)
rub_close_diff = np.diff(rub_close)
snp_close = np.array(SNP.Close[259:])
snp_close_diff = np.diff(snp_close)
```

```python
#Data Dates
date_amzn_close = [0 for i in range(amzn_close.shape[0])]
date_rub_close = [0 for i in range(rub_close.shape[0])]
date_snp_close = [0 for i in range(snp_close.shape[0])]
for i in range(259, AMZN.Date.shape[0]):
    date_amzn_close[i-259] = datetime.strptime(AMZN.Date[i], "%Y-%m-%d")
j=0
for i in range(303, RUB.Date.shape[0]):
    if i-303 in index:
        continue
    date_rub_close[j] = datetime.strptime(RUB.Date[i], "%Y-%m-%d")
    j+=1
for i in range(259, SNP.Date.shape[0]):
    date_snp_close[i-259] = datetime.strptime(SNP.Date[i], "%Y-%m-%d")
```

```python
test_size = 30/1000
```

```python
mdl_arma = arma.ARMA(amzn_close_diff, (8, 0), test_size).fit(exact=False)
mdl_nn = nn.NN(amzn_close_diff, (1, 0, 1), test_size).fit(exact=False)
mdl_armann_zhang = armann_zhang.ARMA_NN_Zhang(amzn_close_diff, (2, 0, 1, 1), test_size).fit(exact=False)
mdl_armann = armann.ARMA_NN(amzn_close_diff, (2, 0, 1, 0, 1), test_size).fit(exact=False)
```

```python
MDLS = [mdl_arma, mdl_nn, mdl_armann_zhang, mdl_armann]
lbl_mdl_armann = "ARIMA-ANN(2, 0, 1, 1, 0, 1)"
lbl_mdl_arma = "ARIMA(8, 1, 0)"
lbl_mdl_nn = "ANN(1, 1, 0, 1)"
lbl_mdl_armann_zhang = "ARIMA-ANN by Zhang(2, 1, 0, 1, 1)"
LABELS = [lbl_mdl_arma, lbl_mdl_nn, lbl_mdl_armann_zhang, lbl_mdl_armann]
label_data = "Цена закрытия на акции Amazon"
xlabel = "Время"
ylabel = "Долларов за акцию" 
#file_path = "RUB/arimann_rub.png"
```

```python
addmet.plot_model_result(amzn_close, mdl_armann, data_index=date_amzn_close, test_index=mdl_armann.test_index, 
                         label_model=lbl_mdl_armann, label_data=label_data, xlabel=xlabel, ylabel=ylabel)
```

```python
addmet.plot_result(amzn_close, MDLS, data_index=date_amzn_close, test_index=mdl_arma.test_index, 
                   LABELS=LABELS, label_data=label_data, xlabel=xlabel, ylabel=ylabel)
```
